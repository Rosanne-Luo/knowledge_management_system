```python
file1 = "/Users/Rosanne/Downloads/投资理财/data/深圳证券交易所.csv"
file2 = "/Users/Rosanne/Downloads/投资理财/data/上海证券交易所.csv"
```

```python
c.execute("drop table stock")
```

```python
c.execute('''CREATE TABLE STOCK
(ID INT PRIMARY KEY NOT NULL,
STOCK_CODE TEXT NOT NULL,
STOCK_NAME TEXT NOT NULL,
MARKET_DATE DATE,
MARKET_ADDRESS TEXT,
ESTABLISHED_DATE DATA,
PROVINCE TEXT,
CITY TEXT,
INDUSTRY_1 TEXT,
INDUSTRY_2 TEXT,
INDUSTRY_3 TEXT
);
''')
```

```python
data = pd.read_csv(file2, encoding="gb18030")
```

```python
for index, row in data.iterrows():
    stock_info = "INSERT INTO STOCK (ID, STOCK_CODE, STOCK_NAME, MARKET_DATE, MARKET_ADDRESS \
    ,ESTABLISHED_DATE, PROVINCE, CITY, INDUSTRY_1, INDUSTRY_2, INDUSTRY_3 ) \
    VALUES (\"{}\",\"{}\",\"{}\",\"{}\",\"{}\",\"{}\",\"{}\",\"{}\",\"{}\",\"{}\",\"{}\")".format(index+2330, str(row["证券代码"]).rjust(6,'0'),row["证券简称"],row["上市日期"], 
         row["上市地点"], row["成立日期"], row["省份"], row["城市"],
         row["申万行业一级名称"], row["申万行业二级名称"], row["申万行业三级名称"])
    print(stock_info)
    c.execute(stock_info)
conn.commit()
```
